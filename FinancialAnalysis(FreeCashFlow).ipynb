```python
path = "C:/Users/Ufuk Altan/Downloads/Downloads/Data Science/PyFinance/04 - Intrinsic Value/"
df = pd.read_csv(path + '50-fcf.csv', index_col = 0)
df.head()
```

```python
df = df.transpose()
```

```python
df.head()
```

```python
df['10Y CAGR'] = (df[2019] / df[2009])**(1/10) -1
df['5Y CAGR'] = (df[2019] / df[2014])**(1/5) -1
df['1Y CAGR'] = (df[2019] / df[2018])**(1/11) -1
```

```python
data = df[['10Y CAGR', '5Y CAGR', '1Y CAGR']]
```
