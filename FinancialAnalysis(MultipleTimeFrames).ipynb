```python
import pandas as pd
```

```python
path = "C:\\Users\\Ufuk Altan\\Downloads\\Downloads\\Data Science\\PyFinance\\08 - Time series stock data\\"
fb = pd.read_csv(path + 'ticker-fb.csv', index_col = 0, parse_dates = True)
google = pd.read_csv(path + 'ticker-goog.csv', index_col = 0, parse_dates = True)
hp = pd.read_csv(path + 'ticker-hp.csv', index_col = 0, parse_dates = True)
nvidia = pd.read_csv(path + 'ticker-nvda.csv', index_col = 0, parse_dates = True)
```

```python
data = pd.concat([fb['Adj Close'], google['Adj Close'], hp['Adj Close'], nvidia['Adj Close']], axis = 1)
data.columns = ['FB', 'GOOGLE', 'HP', 'NVIDIA']
```

```python
data = data.loc['2020-01-02': '2020-12-31', :]
```

```python
daily_change = data / data.iloc[0]
```

```python
data.pct_change()
```

```python
(data.pct_change() + 1).cumprod()
```
