```python
import pandas as pd
```

```python
path = "C:/Users/Ufuk Altan/Downloads/Downloads/Data Science/PyFinance/04 - Intrinsic Value/"
df = pd.read_csv(path + '50-bvps.csv', index_col = 0)
df.head()
```

```python
df.index = pd.to_datetime(df.index)
df.index
```

```python
df = df.resample('Q', convention = 'end').agg('mean')
```

```python
df = df.loc['2010-09-01':]
```

```python
df.head()
```

```python
df = df.transpose()
```

```python
result = pd.DataFrame(data = None, index =df.index)
result['10Y CAGR'] = (df['2020-09-30'] / df['2010-09-30']) ** (1/10) - 1
result['5Y CAGR'] = (df['2020-09-30'] / df['2015-09-30']) ** (1/5) - 1
result['1Y CAGR'] = (df['2020-09-30'] / df['2019-09-30']) ** (1/1) - 1
```
