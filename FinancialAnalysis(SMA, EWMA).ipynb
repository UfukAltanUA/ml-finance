```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "C:\\Users\\Ufuk Altan\\Downloads\\Downloads\\Data Science\\PyFinance\\08 - Time series stock data\\"
fb = pd.read_csv(path + 'ticker-fb.csv', index_col = 0, parse_dates = True)
fb.head()
```

```python
fb['MA10'] = fb['Close'].rolling(10).mean()
fb['MA50'] = fb['Close'].rolling(50).mean()
```

```python
fb.head(11)
```

```python
fb[['Close', 'MA10', 'MA50']].plot(figsize = (15,8))
```

```python
#EWMA
```

```python
fb['EMA50'] = fb['Close'].ewm(span = 50 ,adjust = False).mean()
```
