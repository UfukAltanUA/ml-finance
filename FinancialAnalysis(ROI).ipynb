```python
import pandas as pd
```

```python
path = "C:/Users/Ufuk Altan/Downloads/Downloads/Data Science/PyFinance/04 - Intrinsic Value/"
df = pd.read_csv(path + '50-roi.csv', index_col = 0)
df.head()
```

```python
df.index = pd.to_datetime(df.index)
df.index
```

```python
df.sort_index(ascending = False, inplace = True)
```

```python
df = df.loc[:'2010-01-01']
```

```python
df_T = df.resample('Q', convention = 'end').agg('mean').transpose()
```

```python
df_T['Mean'] = df_T.apply(pd.Series.mean, axis = 1)
```

```python
df_T['Mean'] >= 10
```

```python
df_T.sort_values('Mean', ascending = False)
```
