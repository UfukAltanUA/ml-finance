```python
from yahoofinancials import YahooFinancials

ticker = 'AAPL'
yf = YahooFinancials(ticker)

income_statements = yf.get_financial_stmts('annual', 'income')
```

```python
import pandas as pd
```

```python
income_statements['incomeStatementHistory']['AAPL']
```

```python
df_list = []

for d in income_statements['incomeStatementHistory']['AAPL']:
    df_list.append(pd.DataFrame.from_dict(d, orient = 'index'))
    
IS = pd.concat(df_list)
```
