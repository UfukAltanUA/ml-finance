```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
from datetime import datetime , timedelta
import pandas_datareader as pdr
```

```python
end = datetime.today()
start = end - timedelta(days = 365)
```

```python
tickers = ['VESTL.IS', 'ARCLK.IS', 'BFREN.IS']
df = pdr.get_data_yahoo(tickers, start, end)
```

```python
data = df['Adj Close']
```

```python
X = np.random.randn(5000)
Y = np.random.randn(5000)
```

```python
plt.scatter(X,Y)
```

```python
def regression(ticA, ticB):
    
    X = data[ticA].iloc[1:].values.reshape(-1, 1)
    y = data[ticB].iloc[1:].values.reshape(-1, 1)
    
    regressor = LinearRegression()
    regressor.fit(X,y)
    
    y_pred = regressor.predict(X)
    
    alpha = str(round(regressor.intercept_[0], 5))
    beta = str(round(regressor.coef_[0][0], 5))
    
    
    plt.scatter(X, y)
    plt.plot(X, y_pred, c = 'r')
    plt.title("Alpha: " + alpha + " & Beta: " + beta)
```

```python
regression('VESTL.IS', 'ARCLK.IS')
```

```python
regression('VESTL.IS', 'BFREN.IS')
```
